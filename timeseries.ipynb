1. (2 pts) What is the Nyquist frequency for this recording? Make sure you use the right units.

```python
sampleInterval_s = (time_ms[1] - time_ms[0]) / 1000
sampleFreq_Hz = 1 / sampleInterval_s
nyquistFreq_Hz = sampleFreq_Hz / 2

print(f"Nyquist frequency = {nyquistFreq_Hz :.2f} Hz")
```

2. (2 pts) Properly downsample to every 10th data point. Plot the original recording overlaid with the downsampled version.

```python
from scipy import signal
import matplotlib.pyplot as plt

subtime_ms = time_ms[::10]
subdata = signal.decimate(data, 10)

plt.figure(figsize=(15,5))
plt.plot(time_ms, data, label='original')
plt.plot(subtime_ms, subdata, label='downsampled')
plt.xlabel('Time (ms)')
plt.legend();
```

3. (3 pts) Bandpass filter the recording from 10-30 Hz. Plot the filter kernel and its frequency response. Also plot the original recording overlaid with the filtered version. Also plot the power spectral density of the original recording overlaid with the power spectral density for the filtered version.

```python
bandpass = signal.firwin(201, [10, 30], pass_zero=False, nyq=nyquistFreq_Hz)

plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(bandpass, '.-')
plt.title('bandpass filter')

radiansPerSample, frequencyResponse = signal.freqz(bandpass)
freq_Hz = radiansPerSample / (2 * np.pi) * sampleFreq_Hz

plt.subplot(122)
plt.plot(freq_Hz, np.abs(frequencyResponse), '.-')
plt.xlabel('Frequency (Hz)')
plt.title('frequency response');
```

```python
filteredData = np.convolve(data, bandpass, 'same')

plt.figure(figsize=(15,5))
plt.plot(time_ms, data, label='original')
plt.plot(time_ms, filteredData, label='filtered', linewidth=3)
plt.xlabel('Time (ms)')
plt.legend();
```

4. (3 pts) Filter the recording by convolving it with a Gaussian kernel. Use 51 points for the kernel with the Gaussian centered (i.e. mean at the 26th point) and a standard deviation of 2 pts. Also plot each of your kernels overlaid and labeled with a legend.

```python
import scipy.stats as st

kernel2 = st.norm.pdf(range(-25, 26), 0, 2)
kernel5 = st.norm.pdf(range(-25, 26), 0, 5)
kernel10 = st.norm.pdf(range(-25, 26), 0, 10)

filtered2 = np.convolve(data, kernel2, 'same')
filtered5 = np.convolve(data, kernel5, 'same')
filtered10 = np.convolve(data, kernel10, 'same')

plt.figure(figsize=(15,5))
plt.plot(time_ms, data, label='original')
plt.plot(time_ms, filtered2, lw=2, label='filtered std=2')
plt.plot(time_ms, filtered5, lw=2, label='filtered std=5')
plt.plot(time_ms, filtered10, lw=2, label='filtered std=10')
plt.xlabel('Time (ms)')
plt.legend();
```
